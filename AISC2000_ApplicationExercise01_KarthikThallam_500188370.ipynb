##### All the confirmed positive cases are collected from 34 cities in Ontario

```python
df['Reporting_PHU_City'].value_counts()
```

##### Most no. of positive cases are reported in Toronto followed by  Mississauga, Newmarket, Ottawa where as the least of no. of positive cases are reported in New Liskeard, Sault Ste. Marie, North Bay, and Pembroke

```python
city_count = df['Reporting_PHU_City'].value_counts()

city_count_high = city_count[:12, ]

city_count_low = city_count[30:35, ]

city_count_moderate = city_count[12:30, ]
```

```python
plt.figure(figsize=(20, 8))
plt.title('Cities in Red Zone ')
plt.xlabel('City')
plt.ylabel('No. of positive cases reported')

sns.barplot(city_count_high.index, city_count_high.values, alpha=0.8)
```

```python
plt.figure(figsize=(20, 5))
plt.title('Cities in Green Zone')
plt.xlabel('City')
plt.ylabel('No. of positive cases reported')

sns.barplot(city_count_low.index, city_count_low.values, alpha=0.8)
```

```python
plt.figure(figsize=(20, 5))
plt.title('Cities in Orange Zone')
plt.xlabel('City')
plt.ylabel('No. of positive cases reported')

sns.barplot(city_count_moderate.index, city_count_moderate.values, alpha=0.8)
```

```python
age_df = df['Age_Group'].value_counts()
age_df
```

##### Most of no. of positive cases are reported for the age group of 20s followed by 30s while the least no. of cases were seen in 90s and 80s age groups. 

```python
plt.figure(figsize=(20, 8))
plt.title('Age Groups')
plt.xlabel('Age')
plt.ylabel('No. of positive cases reported')

sns.barplot(age_df.index, age_df.values, alpha=0.8)
```

```python
gender_df = df['Client_Gender'].value_counts()

gender_df
```

```python
plt.figure(figsize=(20, 5))
plt.title('Cases impact on Gender')
plt.xlabel('Gender')
plt.ylabel('No. of positive cases reported')

sns.barplot(gender_df.index, gender_df.values, alpha=0.8)
```

##### There is no big difference in the impact of Covid-19 positive cases on Gender but there are very few cases of Unspecified gender and gender diverse which makes sense when ratio is considered into this matter

```python
status_df = df['Outcome1'].value_counts()

status_df
```

```python
plt.figure(figsize=(20, 5))
plt.title('Current status of the positively reported cases')
plt.xlabel('Satus')
plt.ylabel('No. of positive cases reported')

sns.barplot(status_df.index, status_df.values, alpha=0.8)
```

##### It seems like most of no. of positively tested people are already recovered from covid whereas few people's condition is fatal and some cases are yet to be resolved

```python
plt.figure(figsize=(22,15))

sns.scatterplot(x='Reporting_PHU_Longitude', y='Reporting_PHU_Latitude', data=df, hue='Reporting_PHU_City', palette='tab10')
```

###### So, from the above scatterplot it is clear that more no.of cases are accumulated around -80.0 longitude which is the region of tornto, mississauga and hamilton and interpreted by hue with the colors Brown, Orange amd Pink respectively
